```python
exp = 'STD_1000'
site = 'cdp'

path_in = '/home/lalandmi/eccc/SnowC2-CLASSIC-2D-model-private/inputFiles/SnowMIP/'+site # example file
path_out = '/home/lalandmi/eccc/SnowC2-CLASSIC-2D-model-private/inputFiles/SnowMIP/'+site
```

```python
ds = xr.open_dataset(path_in+'/cdp_init_run_DEF_2D.nc')
ds.load()
```

```python
ds.soilcmas 
```

```python
ds.soilcmas.sum()
```

```python
total = []
for i, fcancmx in enumerate(ds.fcancmx.squeeze().values):
    total.append(fcancmx*ds.soilcmas[0, :, i, 0, 0].sum())

np.sum(total)
```
